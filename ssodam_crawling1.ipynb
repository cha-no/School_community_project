```python
# 필요한 모듈 import

import requests
import re
import time

from urllib.request import urlopen
from urllib import parse
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
# 서담 로그인 코드 200이 나오면 정상 로그인

session=requests.Session()
login = {"id":"본인ID","password":"본인PASSWORD"}
first_page = session.get('http://www.ssodam.com')
html = first_page.text
soup = bs(html, 'html.parser')
csrf = soup.find('input', {'name': 'csrfmiddlewaretoken'})
login={**login,'csrfmiddlewaretoken':csrf['value']}
login_req = session.post('http://www.ssodam.com/auth',data=login)
print(login_req.status_code)
```

```python
page = 1
title = []
head = []
comment_num = []
writer=[]
like = []
upload_date = []
view = []
link = []
```

```python
# 제목 등 기본 게시글 정보 크롤링 코드

while 240-page:
    good = session.get('http://ssodam.com/today/'+str(page))
    time.sleep(0.5)
    soup2 = bs(good.text,'html.parser')
    body = soup2.find('tbody')
    tr = body.find_all('tr')

    for i in range(1,len(tr)):
        s = '$'.join([i.text.strip() for i in tr[i].find_all('span')])
        s = s.split(sep='$')
        if len(s) == 2:
            try:
                comment_num.append(int(re.sub('\D+','',s[1])))
                head.append(None)
                title.append(s[0])
            except:
                head.append(s[0])
                title.append(s[1])
                comment_num.append(0)
        elif len(s)==3:
            head.append(s[0])
            title.append(s[1])
            comment_num.append(int(re.sub('\D+','',s[2])))
        else:
            head.append(None)
            title.append(s[0])
            comment_num.append(0)

        temp = tr[i].find_all('td',style="text-align:center;")
        writer.append(temp[0].text)
        like.append(int(temp[1].text))
        upload_date.append(temp[2].text)
        view.append(int(temp[3].text))
        link.append('http://www.ssodam.com/'+tr[i].find('a')['href'])
        print(page,i)
    page+=1
dic = {'title':title, 'header':head, 'comment_num':comment_num,'writer':writer,'like':like,'upload_date':upload_date,'view':view,'link':link}
df = pd.DataFrame(dic)
```
