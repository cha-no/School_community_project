```python
# 필요한 모듈 import

import requests
import re
import time
import tqdm
from urllib.request import urlopen
from urllib import parse
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
# 서담 로그인 코드 200이 나오면 정상 로그인

session=requests.Session()
login = {"id":"본인ID","password":"본인PASSWORD"}
first_page = session.get('http://www.ssodam.com')
html = first_page.text
soup = bs(html, 'html.parser')
csrf = soup.find('input', {'name': 'csrfmiddlewaretoken'})
login={**login,'csrfmiddlewaretoken':csrf['value']}
login_req = session.post('http://www.ssodam.com/auth',data=login)
print(login_req.status_code)
```

```python
df = pd.read_csv('서담.txt', sep='|')
df
```

```python
links = df['link']
```

```python
links[22]
```

```python
# 링크에 접속해 본문, 댓글등 크롤링

boards = []
texts = []
dislikes = []
comments = []

for i, link in enumerate(links):
    posting = session.get(link)
    soup3 = bs(posting.text,'html.parser')
    
    board = soup3.find('div',{'class':'page-name'}).text.strip()[1:]
    boards.append(board)
    print(i, board)
    if board == '취업게시판' or board == '로스쿨':
        texts.append(soup3.find('div',{'class':'board-content mverdana backlogo'}).text.strip().split('\n')[0])
    else:
        texts.append(soup3.find('div',{'class':'board-content mverdana'}).text)
    
    dislikes.append(soup3.find_all('span',{'class':'badge'})[1].text)
    
    comments.append([[i.text] for i in soup3.find_all('div',{'class':'mverdana comment-content'})])
```

```python
len(boards), len(texts), len(dislikes), len(comments)
```

```python
dic = {'borad':boards, 'text':texts, 'dislike':dislikes,'comment':comments}
df_ssodam = pd.DataFrame(dic)
df_ssodam
```

```python
df_ssodam.to_csv('서담2.txt',sep='|')
```
